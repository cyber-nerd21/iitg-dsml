```python
import pandas as pd  
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sns 

sns.set_style("whitegrid")
```

```python
df = pd.read_csv("Titanic-Dataset.csv")
df.head()
```

```python
df.shape
```

```python
df.info() 
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
sns.countplot(x="Survived", data=df)
plt.title("Survival Count")
plt.show()
```

```python
sns.countplot(x="Survived", hue="Sex" , data = df)
plt.title("Survived by Gender")
plt.show() 
```

```python
sns.countplot(x="Pclass", data=df)
plt.title("Passenger Class Distribution")
plt.show()
```

```python
sns.histplot(df["Age"],  bins = 30  ,  kde = True     )  
plt.title("Age Distribution")
plt.show()
```

```python
plt.figure(figsize = (10,6))
sns.heatmap(df.corr(numeric_only = True ) ,  annot=True , cmap = "coolwarm" )
plt.title("Correlation Heatmap")
plt.show()

```

KEY INSIGHTS   

1. Female passengers had a significantly higher survival rate.
2. First class passengers survived more than third class passengers.
3. Most passengers were between ages 20–40.
4. Passenger class shows correlation with survival probability.
